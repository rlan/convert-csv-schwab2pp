This notebook will download code from Github, verify the example and convert ```schwab.csv```.

1. Run this entire notebook once (Runtime -> Run All). Because ```schwab.csv``` (your Schwab CSV file) doesn't exist yet, there will be an error.
2. Upload your ```schwab.csv```. See instructions below.
